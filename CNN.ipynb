```python
import nibabel as nib
import numpy as np
from PIL import Image

def nii_to_jpg(nii_file, jpg_file):
  """Converts a .nii file to a .jpg file.

  Args:
    nii_file: The path to the .nii file.
    jpg_file: The path to the .jpg file.
  """

  # Load the .nii file.
  img = nib.load(R01-152.nii)

  # Get the image data.
  img_data = img.get_data()

  # Convert the image data to a NumPy array.
  img_array = np.array(img_data)

  # Convert the NumPy array to a PIL image.
  img_pil = Image.fromarray(img_array)

  # Save the PIL image to a .jpg file.
  img_pil.save(jpg_file)

if __name__ == '__main__':
  # Get the path to the .nii file.
  nii_file = 'path/to/nii/file.nii'

  # Get the path to the .jpg file.
  jpg_file = 'path/to/jpg/file.jpg'

  # Convert the .nii file to a .jpg file.
  nii_to_jpg(nii_file, jpg_file)
```

```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def nifti_to_jpg(nifti_path, output_path):
    # Load NIfTI file
    nifti_img = nib.load(nifti_path)
    
    # Extract data from the NIfTI image
    data = nifti_img.get_fdata()

    # Choose a slice to convert to image (you may need to adjust this based on your data)
    slice_index = data.shape[-1] // 2

    # Display the slice using matplotlib
    plt.imshow(data[:, :, slice_index], cmap='gray')
    plt.axis('off')

    # Save the displayed image as a JPEG file
    plt.savefig(output_path, bbox_inches='tight', pad_inches=0)
    plt.close()

def batch_convert_nifti_to_jpg(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Iterate through NIfTI files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".nii"):
            nifti_file_path = os.path.join(input_folder, filename)
            output_jpg_path = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.jpg")
            nifti_to_jpg(nifti_file_path, output_jpg_path)

# Example usage
input_folder_path = 'C:/Users/cshar/Downloads/images/images_original'
output_folder_path = 'C:/Users/cshar/OneDrive/Desktop/major proj'
batch_convert_nifti_to_jpg(input_folder_path, output_folder_path)
```

```python


# Function to read and preprocess images
def read_and_preprocess_images(folder_path):
    images = []
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg')):
            image_path = os.path.join(folder_path, filename)
            img = cv2.imread(image_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
            img = cv2.resize(img, (224, 224))  # Resize images if needed
            images.append(img)
    return np.array(images)

# Load and preprocess images
image_folder = 'images'
images = read_and_preprocess_images(image_folder)

# Flatten the images to be compatible with KMeans
flat_images = images.reshape((images.shape[0], -1))

# Choose the number of clusters (K)
k = 2  # You can adjust this based on the number of classes you want

# Apply K-means clustering
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(flat_images)

# Get labels assigned by K-means
labels = kmeans.labels_

# Visualize some random images with their assigned labels
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
for i in range(6):
    random_index = np.random.randint(len(images))
    axs[i // 3, i % 3].imshow(images[random_index])
    axs[i // 3, i % 3].set_title(f'Cluster {labels[random_index]}')
    axs[i // 3, i % 3].axis('off')
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# Define constants
input_shape = (128, 128, 3)
batch_size = 64
epochs = 20  # Increased number of epochs
learning_rate = 0.001  # Adjust as needed

# Define the CNN model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),  # Added another conv layer
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),  # Increased number of neurons
    #layers.Dropout(0.5),  # Added dropout layer for regularization
    layers.Dense(1, activation='sigmoid')
])

# Compile the model with a specified learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Create data generators with aggressive data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=20,  # Added rotation augmentation
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    'train',
    target_size=(128, 128),
    batch_size=batch_size,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    'test',
    target_size=(128, 128),
    batch_size=batch_size,
    class_mode='binary'
)

# Implement early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model with augmented data and shuffle during training
history = model.fit(train_generator, epochs=epochs, validation_data=test_generator, callbacks=[early_stopping])

# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
